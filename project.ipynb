```python
#10
filtered_data1 = data1[data1.Deaths > 150]
filtered_data2 = data2[data2.Deaths > 150]

x1 = filtered_data1['Country_Region']
yd1 = filtered_data1['Deaths']
yc1 = filtered_data1['Confirmed']
yr1 = filtered_data1['Recovered']
ya1 = filtered_data1['Active']

x2 = filtered_data2['Country_Region']
yd2 = filtered_data2['Deaths']
yc2 = filtered_data2['Confirmed']
yr2 = filtered_data2['Recovered']
ya2 = filtered_data2['Active']

plt.figure(figsize=(30, 10))
plt.subplot(121)
plt.plot(x1, yd1, color='blue', linewidth=4)
plt.plot(x1, yc1, color='red', linewidth=4, alpha=0.4)
plt.plot(x1, yr1, color='green', linewidth=4, alpha=0.7)
plt.plot(x1, ya1, color='yellow', linewidth=4)
plt.xticks(rotation = 90)
plt.title('1: Total Deaths(>150),Confirmed,Recovered and Active Cases by Country')

plt.figure(figsize=(30, 10))
plt.subplot(221)
plt.plot(x2, yd2, color='blue', linewidth=4)
plt.plot(x2, yc2, color='red', linewidth=4, alpha=0.4)
plt.plot(x2, yr2, color='green', linewidth=4, alpha=0.7)
plt.plot(x2, ya2, color='yellow', linewidth=4)
plt.xticks(rotation = 90)
plt.title('2: Total Deaths(>150),Confirmed,Recovered and Active Cases by Country')


plt.show()
```
